# echartsy Examples

A collection of examples ranging from simple to complex, demonstrating
the full power of echartsy's matplotlib-style API for Apache ECharts.

---

```python
import echartsy as ec
import pandas as pd
import numpy as np

# Configure rendering engine
# Options: "jupyter" (inline widget), "python" (browser popup), "streamlit"
ec.config(engine="jupyter")

# ec.Figure() and ec.figure() are identical — figure() is just a shorthand factory
print(f"echartsy v{ec.__version__} ready!")
```

---
## 1. Simple Bar Chart

The most basic chart — just pass a DataFrame with x and y columns.

```python
df_fruit = pd.DataFrame({
    "Fruit": ["Apples", "Bananas", "Cherries", "Dates", "Elderberries"],
    "Sales": [120, 95, 78, 42, 63]
})

fig = ec.figure(renderer='svg')
fig.bar(df_fruit, x="Fruit", y="Sales")
fig.title("Fruit Sales")
fig.pie(df=df_fruit,names='Fruit',values='Sales',center=['85%','25%'],radius=['10%', '30%'], label_font_size=5)
fig.legend(top=40)
fig.margins(top=40)
fig.to_html("revenue_report.html")
fig.toolbox()
fig.show()
```

---
## 2. Simple Line Chart

Use `.plot()` just like `plt.plot()` — with optional smoothing.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df_temp = pd.DataFrame({
    "Month": months,
    "Temperature": [5, 7, 12, 18, 23, 28, 31, 30, 25, 18, 11, 6]
})

fig = ec.figure(height="400px")
fig.plot(df_temp, x="Month", y="Temperature", smooth=True)
fig.title("Average Monthly Temperature")
fig.ylabel("Temp (°C)")
fig.show()
```

---
## 3. Pie / Donut Chart

Pie charts use `names` and `values` instead of x/y.

```python
df_market = pd.DataFrame({
    "Browser": ["Chrome", "Safari", "Firefox", "Edge", "Other"],
    "Share": [65.7, 18.2, 6.3, 5.1, 4.7]
})

fig = ec.figure(height="450px")
fig.pie(df_market, names="Browser", values="Share", inner_radius="40%")
fig.title("Browser Market Share 2025")
fig.show()
```

---
## 4. Scatter Plot

Scatter plots with optional `color` and `size` encoding.

```python
np.random.seed(42)
n = 80
df_scatter = pd.DataFrame({
    "Height": np.random.normal(170, 10, n),
    "Weight": np.random.normal(70, 12, n),
    "Gender": np.random.choice(["Male", "Female"], n)
})

fig = ec.figure(height="450px")
fig.scatter(df_scatter, x="Height", y="Weight", color="Gender")
fig.title("Height vs Weight")
fig.xlabel("Height (cm)")
fig.ylabel("Weight (kg)")
fig.legend(top=40)
fig.show()
```

## 5. Grouped Bar Chart with `hue`

Pass a `hue` column to automatically create grouped (side-by-side) bars.

```python
df_revenue = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"] * 3,
    "Revenue": [120, 145, 160, 180, 90, 110, 130, 155, 60, 80, 95, 110],
    "Region": ["North"] * 4 + ["South"] * 4 + ["West"] * 4
})

fig = ec.figure(height="450px")
fig.bar(df_revenue, x="Quarter", y="Revenue", hue="Region")
fig.title("Quarterly Revenue by Region")
fig.ylabel("Revenue ($K)")
fig.legend(top=40)
fig.show()
```

---
## 6. Stacked Bar Chart

Combine `hue` with `stack=True` to stack the bars.

```python
fig = ec.figure(height="450px")
fig.bar(df_revenue, x="Quarter", y="Revenue", hue="Region",
        stack=True, labels=True)
fig.palette(ec.PALETTE_RUSTY)
fig.title("Stacked Revenue by Region")
fig.ylabel("Revenue ($K)")
fig.legend(top=40)
fig.show()
```

## 7. Histogram

## 8. Area Chart

Line charts become area charts with `area=True`.

```python
df_traffic = pd.DataFrame({
    "Hour": list(range(24)),
    "Users": [12, 8, 5, 3, 4, 8, 25, 60, 95, 120, 110, 105,
              115, 108, 100, 98, 112, 130, 95, 70, 50, 35, 22, 15],
})

fig = ec.figure(height="400px")
fig.plot(df_traffic, x="Hour", y="Users", smooth=True,
         area=True, area_opacity=0.3)
fig.title("Website Traffic — 24hr")
fig.xlabel("Hour of Day")
fig.ylabel("Active Users")
fig.legend(top=40)
fig.show()
```

## 9. Horizontal Bar Chart

Set `orient="h"` for horizontal bars — great for long category names.

## 10. Multi-series: Bar + Line Overlay (Dual Axis)

Combine bar and line on the same chart using `axis=1` for the secondary y-axis.

```python
df_sales = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [250, 310, 280, 400, 520, 480],
})

df_growth = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "GrowthRate": [5.2, 8.1, 6.5, 12.3, 18.7, 15.0],
})

fig = ec.figure(height="500px")
fig.bar(df_sales, x="Month", y="Revenue", labels=True)
fig.plot(df_growth, x="Month", y="GrowthRate", smooth=True, axis=1)
fig.title("Revenue & Growth Rate")
fig.ylabel("Revenue ($K)")
fig.legend(top=40)
fig.ylabel_right("Growth %")
fig.show()
```

## 11. Radar Chart

Compare multi-dimensional profiles side by side.

```python
indicators = [
    {"name": "Attack", "max": 100},
    {"name": "Defense", "max": 100},
    {"name": "Speed", "max": 100},
    {"name": "Stamina", "max": 100},
    {"name": "Magic", "max": 100},
    {"name": "Intelligence", "max": 100},
]

data = [
    [90, 60, 80, 70, 40, 55],   # Warrior
    [30, 40, 50, 60, 95, 90],   # Mage
    [70, 50, 95, 80, 20, 65],   # Rogue
]

fig = ec.figure(height="500px")
fig.radar(indicators, data, series_names=["Warrior", "Mage", "Rogue"])
fig.title("Character Class Comparison")
fig.palette(ec.PALETTE_CLINICAL)
fig.show()
```

## 12. Heatmap

Visualize a matrix of values — great for correlations or schedules.

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
hours = ["9am", "10am", "11am", "12pm", "1pm", "2pm", "3pm", "4pm"]

np.random.seed(123)
rows = []
for d in days:
    for h in hours:
        rows.append({"Day": d, "Hour": h, "Emails": np.random.randint(2, 50)})

df_heatmap = pd.DataFrame(rows)

fig = ec.figure(height="450px")
fig.heatmap(df_heatmap, x="Hour", y="Day", value="Emails")
fig.title("Emails Received per Hour")
fig.show()
```

## 14. Treemap

Hierarchical data as nested rectangles. Pass `path` as a list of columns
from broadest to most specific.

```python
df_tree = pd.DataFrame({
    "Category": ["Electronics", "Electronics", "Electronics",
                 "Clothing", "Clothing", "Clothing",
                 "Food", "Food"],
    "SubCategory": ["Phones", "Laptops", "Tablets",
                    "Shirts", "Pants", "Shoes",
                    "Fruit", "Dairy"],
    "Sales": [500, 420, 180, 300, 250, 200, 150, 100]
})

fig = ec.figure(height="500px")
fig.treemap(df_tree, path=["Category", "SubCategory"], value="Sales", roam=False)
fig.title("Sales by Category")
fig.show()
```

## 15. Sankey Diagram

Visualize flow between stages. `levels` defines the columns in order.

```python
df_flow = pd.DataFrame({
    "Source": ["Organic", "Organic", "Paid", "Paid", "Referral", "Referral"],
    "Channel": ["Landing Page A", "Landing Page B",
                "Landing Page A", "Landing Page B",
                "Landing Page A", "Landing Page B"],
    "Outcome": ["Converted", "Bounced", "Converted", "Bounced",
                "Converted", "Bounced"],
    "Users": [350, 150, 200, 300, 180, 70]
})

fig = ec.figure(height="500px")
fig.sankey(df_flow, levels=["Source", "Channel", "Outcome"], value="Users",)
fig.title("User Flow: Source → Page → Outcome")
fig.show()
```

## 20. Full Dashboard: Bar + Line + Toolbox + DataZoom

Combine everything: dual-axis, toolbox (save as image), and data zoom slider.

```python
np.random.seed(2026)
months_full = pd.date_range("2024-01", periods=24, freq="MS").strftime("%b %Y").tolist()

df_kpi = pd.DataFrame({
    "Month": months_full,
    "Revenue": np.cumsum(np.random.normal(50, 15, 24)).clip(100).tolist(),
})

df_margin = pd.DataFrame({
    "Month": months_full,
    "Margin": (np.random.uniform(15, 35, 24)).tolist(),
})

fig = ec.figure(height="550px", style=ec.StylePreset.KPI_REPORT)
fig.bar(df_kpi, x="Month", y="Revenue", labels=False, border_radius=4)
fig.plot(df_margin, x="Month", y="Margin", smooth=True, axis=1,
         line_width=3, symbol_size=8)
fig.title("KPI Dashboard: Revenue & Margin", subtitle="Jan 2024 – Dec 2025")
fig.ylabel("Cumulative Revenue ($K)")
fig.ylabel_right("Margin %")
fig.toolbox()
fig.legend(top=50)
fig.datazoom()
fig.margins(top=40)
fig.save(name="kpi_dashboard")
fig.show()
```

## 21. Animated Timeline — Bar Race

Use `TimelineFigure` to create animated charts with a playback slider.
Each unique value in `time_col` becomes an animation frame.

```python
np.random.seed(1)
years = ["2020", "2021", "2022", "2023", "2024"]
countries = ["USA", "China", "Germany", "Japan", "India"]

rows = []
for i, yr in enumerate(years):
    for c in countries:
        base = {"USA": 21, "China": 15, "Germany": 4, "Japan": 5, "India": 3}[c]
        gdp = base + i * np.random.uniform(0.3, 1.5) + np.random.normal(0, 0.5)
        rows.append({"Year": yr, "Country": c, "GDP": round(gdp, 2)})

df_gdp = pd.DataFrame(rows)

fig = ec.TimelineFigure(height="500px", interval=1.5)
fig.bar(df_gdp, x="Country", y="GDP", time_col="Year", labels=True)
fig.title("GDP by Country")
fig.ylabel("GDP (Trillion USD)")
fig.legend(top=30)
fig.show()
```

## Composite Charts
The following examples show how to combine multiple chart types in a
single figure. Simply call `.pie()` with a `center` and `radius` on
any existing figure to overlay a pie chart — no escape hatches needed.

For advanced cases, `.raw_series({...})` is still available to inject
any arbitrary ECharts series config.

---
## 25. Bar Chart with Pie Overlay

A small pie chart positioned in the corner of a bar chart, showing
the proportional breakdown while the bars show absolute values.

```python
df_dept = pd.DataFrame({
    "Department": ["Engineering", "Marketing", "Sales", "Design", "Support"],
    "Budget": [450, 280, 320, 180, 120],
})

ec.config(engine='jupyter',adaptive='auto')
# Build a bar chart, then overlay a mini pie in the top-right corner
fig = ec.figure(height="500px", style=ec.StylePreset.MINIMAL)
fig.bar(df_dept, x="Department", y="Budget", border_radius=6,
        gradient=True, gradient_colors=["#f4f1de","#e07a5f"], 
        labels=True,label_color='orange', 
        emphasis=Emphasis(focus='self',
                          item_style=ItemStyle(,)))
fig.title("Department Budgets ($K)")
fig.ylabel("Budget ($K)")
fig.pie(df_dept, names="Department", values="Budget",
        center=["82%", "25%"], radius=["18%", "28%"],
        border_radius=4, label_font_size=9,)
fig.legend(top=40, left=200)
fig.margins(right=120)
fig.show()
```

## 26. Bar + Line + Pie (Triple Composite)

Revenue bars on the left axis, growth line on the right axis, and a
donut showing the product mix — all in one figure.

```python
df_monthly = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [250, 310, 280, 400, 520, 480],
})
df_growth = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Growth": [5.2, 8.1, 6.5, 12.3, 18.7, 15.0],
})

df_mix = pd.DataFrame({
    "Plan": ["Enterprise", "Pro", "Starter", "Free-trial"],
    "Share": [45, 30, 15, 10],
})

fig = ec.figure(height="550px")
# Bar: revenue on left axis
fig.bar(df_monthly, x="Month", y="Revenue", labels=True, border_radius=4)
# Line: growth % on right axis
fig.plot(df_growth, x="Month", y="Growth", smooth=True, axis=1,
         line_width=3, symbol_size=8)
# Pie: product mix overlay in top-right
fig.pie(df_mix, names="Plan", values="Share",
        center=["25%", "32%"], radius=["15%", "25%"],
        border_radius=4, label_font_size=9)
fig.title("Monthly KPIs: Revenue, Growth & Product Mix")
fig.ylabel("Revenue ($K)")
fig.ylabel_right("Growth %")
fig.legend(top=40, left=350)
fig.margins(right=20)
fig.to_html(filepath='./Monthly KPIS Revenue Growth Product Mix.html')
fig.show()
```

## 27. Side-by-Side Pies (Multi-Pie Layout)

Two pie charts in one figure, each positioned at different `center` points.

```python
df_q1 = pd.DataFrame({
    "Product": ["Product A", "Product B", "Product C"],
    "Revenue": [40, 35, 25],
})
df_q2 = pd.DataFrame({
    "Product": ["Product A", "Product B", "Product C"],
    "Revenue": [30, 42, 28],
})

fig = ec.figure(height="450px")
fig.pie(df_q1, names="Product", values="Revenue",
        center=["28%", "55%"], radius=["25%", "40%"],
        border_radius=6, label_font_size=10)
fig.pie(df_q2, names="Product", values="Revenue",
        center=["72%", "55%"], radius=["25%", "40%"],
        border_radius=6, label_font_size=10)
fig.title("Product Mix: Q1 vs Q2")
fig.extra(
    legend={"data": ["Product A", "Product B", "Product C"],
            "top": 40, "left": "center"},
    graphic=[
        {"type": "text", "left": "18%",  "top": "88%",
         "style": {"text": "Q1 2025", "fontSize": 14, "fontWeight": "bold",
                    "fill": "#888"}},
        {"type": "text", "left": "62%", "top": "88%",
         "style": {"text": "Q2 2025", "fontSize": 14, "fontWeight": "bold",
                    "fill": "#888"}},
    ],
)
fig.show()
```

## 28. Stacked Bar with Line Trend + Pie Breakdown

A common dashboard pattern: stacked bars by region, a total-trend line,
and a pie summarizing the overall split.

```python
df_rev = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"] * 3,
    "Revenue": [120, 145, 160, 180, 90, 110, 130, 155, 60, 80, 95, 110],
    "Region":  ["North"] * 4 + ["South"] * 4 + ["West"] * 4,
})

# Compute totals for the trend line
df_totals = df_rev.groupby("Quarter", sort=False)["Revenue"].sum().reset_index()
df_totals.columns = ["Quarter", "Total"]

# Compute region totals for the pie
region_totals = (df_rev.groupby("Region")["Revenue"].sum()
                 .reset_index()
                 .rename(columns={"Revenue": "Total"}))

fig = ec.figure(height="550px")
# Stacked bars by region
fig.bar(df_rev, x="Quarter", y="Revenue", hue="Region",
        stack=True, border_radius=4)
# Total trend line on right axis
fig.plot(df_totals, x="Quarter", y="Total", smooth=True, axis=1,
         line_width=3, symbol_size=10, labels=True)
# Mini pie showing region split
fig.pie(region_totals, names="Region", values="Total",
        center=["26%", "20%"], radius=["12%", "22%"],
        border_radius=4, label_font_size=9)
fig.title("Quarterly Revenue: Stacked + Trend + Breakdown")
fig.ylabel("Revenue ($K)")
fig.ylabel_right("Total ($K)")
fig.margins(right=120)
fig.legend(top=35)
fig.show()
```

## Emphasis Examples

The `emphasis` parameter lets you control hover/highlight behavior using typed Python attributes instead of raw dictionaries.

```python
from echartsy import Emphasis, ItemStyle

df_heat = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"] * 3,
    "Hour": ["9am"] * 5 + ["12pm"] * 5 + ["5pm"] * 5,
    "Value": [10, 20, 30, 25, 15, 35, 40, 50, 45, 30, 20, 25, 35, 30, 10],
})

(
    ec.Figure(height="400px")
    .heatmap(
        df_heat, x="Day", y="Hour", value="Value",
        emphasis=Emphasis(
            item_style=ItemStyle(border_color="#000", border_width=2, shadow_blur=15),
        ),
    )
    .title("Heatmap — Custom Hover Border")
    .show()
)
```

```python
from echartsy.emphasis import ScatterEmphasis

df_scatter = pd.DataFrame({
    "Height": [160, 165, 170, 175, 180, 185, 155, 172, 168, 190],
    "Weight": [55, 62, 68, 72, 78, 85, 50, 70, 65, 90],
})

(
    ec.Figure(height="400px")
    .scatter(
        df_scatter, x="Height", y="Weight",
        emphasis=ScatterEmphasis(disabled=False),
    )
    .title("Scatter — Emphasis Disabled")
    .xlim(150,200)
    .show()
)
```

```python
from echartsy.emphasis import PieEmphasis, LabelStyle

df_pie = pd.DataFrame({
    "Category": ["Electronics", "Clothing", "Food", "Books"],
    "Amount": [450, 300, 250, 150],
})

(
    ec.Figure(height="400px")
    .pie(
        df_pie, names="Category", values="Amount",
        emphasis=PieEmphasis(
            scale=True,
            scale_size=15,
            item_style=ItemStyle(shadow_blur=20, shadow_color="rgba(0,0,0,0.6)"),
            label=LabelStyle(show=True, font_size=16, font_weight="bold"),
        ),
    )
    .title("Pie Emphasis — Scale on Hover")
    .show()
)
```

```python
from echartsy.emphasis import LineEmphasis, LineStyle, LabelStyle

df_line = pd.DataFrame({
    "Date": ["Q1", "Q2", "Q3", "Q4"] * 2,
    "Price": [150, 180, 170, 200, 90, 110, 130, 120],
    "Stock": ["AAPL"] * 4 + ["GOOG"] * 4,
})

(
    ec.Figure(height="400px")
    .plot(
        df_line, x="Date", y="Price", hue="Stock",
        emphasis=LineEmphasis(
            focus="series",
            line_style=LineStyle(width=4),
            label=LabelStyle(show=True, formatter="{c}"),
        ),
    )
    .title("Line Emphasis — Bold on Hover")
    .show()
)
```

```python
from echartsy.emphasis import Emphasis, ItemStyle

df_bar = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"] * 2,
    "Revenue": [100, 120, 140, 130, 80, 90, 110, 100],
    "Region": ["East"] * 4 + ["West"] * 4,
})

(
    ec.Figure(height="400px")
    .bar(
        df_bar, x="Month", y="Revenue", hue="Region",
        emphasis=Emphasis(
            focus="series",
            item_style=ItemStyle(shadow_blur=10, shadow_color="rgba(0,0,0,0.3)"),
        ),
    )
    .title("Bar Emphasis — Focus on Series")
    .legend(top=50)
    .show()
)
```
